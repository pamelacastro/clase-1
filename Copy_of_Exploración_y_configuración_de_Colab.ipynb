# Leer una imagen con OpenCV

## Cargar archivos locales en Colab

```python
# prompt: Crea un codigo para leer una imagen con OpenCV y graficar la imagen original junto con sus canales de color RGB con Matplotlib

import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
image = cv2.imread('/content/arbol.jpeg')  # Reemplaza 'image.jpg' con el nombre de tu archivo

# Convertir la imagen de BGR a RGB (OpenCV lee im√°genes en formato BGR)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Obtener los canales de color
red_channel = image_rgb[:, :, 0]
green_channel = image_rgb[:, :, 1]
blue_channel = image_rgb[:, :, 2]

# Mostrar la imagen original y sus canales de color usando Matplotlib
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title('Imagen original')

plt.subplot(2, 2, 2)
plt.imshow(red_channel, cmap='gray')
plt.title('Canal Rojo')

plt.subplot(2, 2, 3)
plt.imshow(green_channel, cmap='gray')
plt.title('Canal Verde')

plt.subplot(2, 2, 4)
plt.imshow(blue_channel, cmap='gray')
plt.title('Canal Azul')

plt.tight_layout()
plt.show()
```

```python
# prompt: de la imagen anterior detecta las partes del arbol como tronco, ramas y hojas

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
image = cv2.imread('/content/arbol.jpeg')  # Reemplaza 'image.jpg' con el nombre de tu archivo
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convertir a espacio de color HSV
hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)

# Definir rangos de color para hojas (verde)
lower_green = np.array([35, 40, 40])
upper_green = np.array([85, 255, 255])

# Crear una m√°scara para las hojas
mask_hojas = cv2.inRange(hsv, lower_green, upper_green)

# Eliminar ruido de la m√°scara
kernel = np.ones((5,5),np.uint8)
mask_hojas = cv2.morphologyEx(mask_hojas, cv2.MORPH_OPEN, kernel)
mask_hojas = cv2.morphologyEx(mask_hojas, cv2.MORPH_CLOSE, kernel)


# Encontrar contornos de las hojas
contours_hojas, _ = cv2.findContours(mask_hojas, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibujar contornos de hojas en la imagen original
image_hojas = image_rgb.copy()
cv2.drawContours(image_hojas, contours_hojas, -1, (0, 255, 0), 3)


# --- Detecci√≥n del tronco (ejemplo b√°sico, se puede mejorar) ---
# Convertir a escala de grises
gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Aplicar un filtro gaussiano para reducir ruido
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Detectar bordes con Canny
edges = cv2.Canny(blurred, 50, 150)

# Encontrar contornos de los objetos
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Buscar el contorno m√°s grande (probable tronco)
trunk_contour = max(contours, key=cv2.contourArea)

# Dibujar el contorno del tronco en la imagen
image_tronco = image_rgb.copy()
cv2.drawContours(image_tronco, [trunk_contour], -1, (0, 0, 255), 3)


# Mostrar resultados
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title('Imagen original')

plt.subplot(2, 2, 2)
plt.imshow(image_hojas)
plt.title('Hojas detectadas')

plt.subplot(2, 2, 3)
plt.imshow(image_tronco)
plt.title('Tronco detectado')

plt.subplot(2, 2, 4)
plt.imshow(mask_hojas, cmap='gray')
plt.title('M√°scara de hojas')

plt.tight_layout()
plt.show()
```

```python
# prompt: operaciones matriciales

import numpy as np

# Ejemplo de matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Suma de matrices
sum_matrix = matrix1 + matrix2
print("Suma de matrices:\n", sum_matrix)

# Resta de matrices
diff_matrix = matrix1 - matrix2
print("Resta de matrices:\n", diff_matrix)

# Multiplicaci√≥n de matrices
product_matrix = np.dot(matrix1, matrix2)  # o matrix1 @ matrix2
print("Producto de matrices:\n", product_matrix)

# Multiplicaci√≥n elemento a elemento
elementwise_product = matrix1 * matrix2
print("Producto elemento a elemento:\n", elementwise_product)

# Transpuesta de una matriz
transposed_matrix = matrix1.T
print("Transpuesta de la matriz:\n", transposed_matrix)

# Inversa de una matriz
try:
    inverse_matrix = np.linalg.inv(matrix1)
    print("Inversa de la matriz:\n", inverse_matrix)
except np.linalg.LinAlgError:
    print("La matriz no es invertible.")

# Determinante de una matriz
determinant = np.linalg.det(matrix1)
print("Determinante de la matriz:\n", determinant)
```
